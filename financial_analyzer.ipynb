```python
import pandas as pd

# Function to clean and process transaction data
def clean_data(df):
    # Drop rows with missing important values
    df = df.dropna(subset=['transaction_status', 'timestamp', 'amount', 'merchant_category'])

    # Keep only successful transactions
    df = df[df['transaction_status'] == 'SUCCESS']

    # Convert timestamp to datetime
    df['datetime'] = pd.to_datetime(df['timestamp'], errors='coerce', dayfirst=True)
    df = df.dropna(subset=['datetime'])

    # Define essential categories
    essential_categories = {'Grocery', 'Utilities', 'Healthcare', 'Education', 'Fuel', 'Transport'}

    # Tag each transaction as essential or non-essential
    df['expense_type'] = df['merchant_category'].apply(
        lambda x: 'essential' if x in essential_categories else 'non-essential'
    )

    # Identify if transaction occurred on a weekend
    df['is_weekend'] = df['datetime'].dt.weekday >= 5

    return df

df = clean_data(df)

# Preview cleaned data
df.head()
```

```python
# Spending Pattern Analysis:

# Calculate total spend per category, essential vs non-essential split, and average transaction amount.

def spending_patterns(df):
    total_spend = df['amount'].sum()
    spend_by_category = df.groupby('merchant_category')['amount'].sum().sort_values(ascending=False)
    spend_essential = df[df['expense_type'] == 'essential']['amount'].sum()
    spend_non_essential = df[df['expense_type'] == 'non-essential']['amount'].sum()
    avg_transaction = df['amount'].mean()

    return {
        "total_spend": total_spend,
        "spend_by_category": spend_by_category,
        "spend_essential": spend_essential,
        "spend_non_essential": spend_non_essential,
        "avg_transaction": avg_transaction
    }

patterns = spending_patterns(df)
print(patterns)
```

```python
# Group by both merchant_category and expense_type
category_expense_summary = df.groupby(['merchant_category', 'expense_type'])['amount'].sum().unstack(fill_value=0)

# Optional: add a total column and sort
category_expense_summary['Total'] = category_expense_summary.sum(axis=1)
category_expense_summary = category_expense_summary.sort_values('Total', ascending=False)

# Show summary
print(category_expense_summary)
```

```python
# Time-based Trends

# Analyze spending by hour, day of week, and monthly spend.
def time_trends(df):
    df['month'] = df['datetime'].dt.to_period('M')

    spend_by_hour = df.groupby('hour_of_day')['amount'].sum()
    spend_by_day = df.groupby('day_of_week')['amount'].sum()
    spend_by_month = df.groupby('month')['amount'].sum()

    return {
        "spend_by_hour": spend_by_hour,
        "spend_by_day": spend_by_day,
        "spend_by_month": spend_by_month
    }

# Call the function
trends = time_trends(df)

# Print each trend section neatly
print("=== Spend by Hour ===")
print(trends['spend_by_hour'])


```

```python
print("\n=== Spend by Day of Week ===")
print(trends['spend_by_day'])
print("\n=== Spend by Month ===")
print(trends['spend_by_month'])
```

```python
# Category-wise Summaries

# Summarize spend, count, and average per category.
def category_summary(df):
    summary = df.groupby('merchant_category').agg(
        total_spend=('amount', 'sum'),
        transaction_count=('amount', 'count'),
        avg_spend=('amount', 'mean')
    ).sort_values('total_spend', ascending=False)
    return summary

cat_summary = category_summary(df)
print(cat_summary)

```

```python
#  Wasteful Transaction Detection 

# Define "wasteful" as high non-essential spend transactions above a threshold.

def detect_wasteful_transactions(df, threshold=1000):
    # Example: non-essential transactions > threshold 
    wasteful = df[(df['expense_type'] == 'non-essential') & (df['amount'] > threshold)]
    return wasteful

wasteful_txns = detect_wasteful_transactions(df)
print(wasteful_txns[['transaction_id', 'merchant_category', 'amount', 'expense_type']])
```
